```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
dataframes = []
names = ['small2', 'small4', 'big2', 'big4']
for name in names:
  ar = []
  for j in range(5):
    df = pd.read_csv(f'csv/{name}_count_q_is_{j}.csv', sep=';')
    ar.append(df)
  dataframes.append(ar)
```

```python
messages = [
    '10000-2',
    '10000-4',
    '100000-2',
    '100000-4'
] 
```

```python
def draw_graphix(df, type_dataset):
  all_algos = set(df['algo_name'].values)
  fig, ax = plt.subplots()
  for algo in all_algos:
    df_sort = df.loc[df['algo_name'] == algo]
    X = df_sort['pattern_len']
    Y = df_sort['time']
    ax.plot(X, Y, label=algo)
  ax.legend()
  fig.set_figheight(15)
  fig.set_figwidth(24)
  message = messages[type_dataset]
  title = message
  ax.set_title(repr(title))
  plt.savefig(f'{title}.png')
```

```python
draw_graphix(dataframes[-1][0], 3)
```
